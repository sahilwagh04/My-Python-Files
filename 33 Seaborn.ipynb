## Bar plot

Gives the estimate of an central tendancy of a numeric variable with a categorical column on x axis and numeric column on y axis

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# way 1
bar =sns.barplot(cars['cyl'],cars['mpg'])
```

```python
# way 2
bar = sns.barplot(x='cyl',y='mpg', data=cars)
```

## distplot

Creates the distribution of any continous data lets create the distribution of miles per models of the different vehicles of the cars the mpg metrics measures the total distance the car can travel per model of the car

```python
sns.distplot(cars.mpg, bins=10,color='g')
```

## heatmap()

```python
import pandas as pd
cars.corr(method='pearson',numeric_only=True)
```

```python
sns.heatmap(cars.corr(method='pearson',numeric_only=True),linewidths=0.5,annot=True)
```

Lighter color have positive correlation and the brighter color have negative correlation
