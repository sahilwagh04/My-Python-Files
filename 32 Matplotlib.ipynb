## Matplotlib Markers
In this section we see more about Markers,

We can use the keyword argument marker to emphasize each point with a specified marker

Eg:

Mark each point with a circle.

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker = 'o')
plt.show()
```

We can give any symbol inside the marker keyword,
Eg:
We can give *.

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker = '*')
plt.show()
#This way we can change the marker and the output will be
```

## Format Strings fmt

You can use also use the shortcut string notation parameter to specify the marker.

This parameter is also called ft, and is written with this syntax:
Marker|line color.

Eg: Mark each point with a circle.

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,'*:r')
plt.show()

#This will be like that string range giving, so like that we can simlpy tell the python do give a plot with line as this, 
#marker as like that.
```

The marker value can be anything from the Marker Reference above.
The line value can be one of the following:

Line Syntax   Description

- '-'         Solid line
- ':'         Dotted line
- '--'        Dashed line
- '-.'        Dashed/dotted line

If you leave out the line value in the fmt parameter, no line will be plotted.

## Marker Size

    You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:
    Eg:
    Set the size of the marker to 20.

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

## Marker Color
    You can use the keyword argument markeredgecolor or the shorter me to set the color of the edge of the markers:
    Eg: set the edge colour to red.

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

### We can change the marker face colour too, by an keyword markerfacecolour or the shorter mfc.

```python
ypoint = np.array([3,8,1,10])
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
#Now we can see that we specify both edge and face will be in red
```

## Matplotlib Labels and Title

- With Pyplot, we can use the xlabel() and ylabel() function to set a label for the x-axis and y-axis.

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x, y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

## Create a Title for a plot
- With Pyplot, you can use the title() function to set a titlefor the plot.
- Eg:

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

## Set Font Properties for Title and Labels

- We can use thr fontdict parameter in xlabel(), ylabel() and title() to set font properties for the title and labels.

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

font1= {'family':'serif','color':'blue','size':20}
font2= {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

## Position the Title
- We can use the loc parameter in title() to position the title.
- Legal values are:'left','right',and 'center'. Default value is 'center'.

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("calorie Burnage")

plt.plot(x, y)
plt.show()
```

## Matplotlib Adding Grid Lines

With Pyplot, you can use the grid() function to add grid lines to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x, y)
plt.grid()
plt.show()
```

We can use the axis parameter in the grid() function to specify which grid lines to display.

Legal values are: 'x', 'y', and 'both'. Default value is 'both'.

```python
# Eg:

#Display only grid lines for the x-axis.

x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x, y)
plt.grid(axis = 'x')
plt.show()
```

## Properties of grid

We can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).

```python
#Eg:
 #set the properties of grid.
    
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x, y)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

## Matplotlib Subplots

What is we need more than one graph, in a single plot.

Which means, subplots. 

We can use subplots() function.

Eg:

```python
# plot 1:
x = np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
```

```python
#Plot 2:

x = np.array([0,1,2,3])
y = ([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x, y)
plt.show()
```

plt.subplot (1, 2, 1) 

plt. subplot (1, 2, 2)
This two lines here represents the figure and the plots in the figure, which means this 2 lines will decide how many subplots are there and where it need to allocate.

1,2,1 = for 1st subplot

1,2,2 = for 2nd subplot

If we type 1,2,3 there will an error, because It calculated like this:1 * 2 = 2

Maximum subplot will be 2

The subplots() Function

The subplots() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

So, if we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side), we can write the syntax like this:

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.plot(x, y)
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.plot(x, y)
x = np.array([0,1,2,3])
y= np.array([3,8,1,10])
plt.subplot(2,3,3)
plt.plot(x, y)
```
