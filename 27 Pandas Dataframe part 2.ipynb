## Conditional Changes

```python
details = {
    'Name' : ['Ankit', 'Aishwarya', 'Ankit', 'Sahil', 'Shaurya', 'Shivangi'],
    'Maths' : [80, 40, 50, 70, 42, 61],
    'Sci' : [80, 21, 43, 61, 72, 71],
    'Hindi' : [53, 21, 73, 78, 92, 11],
    'Chem' : [73, 41, 23, 89, 45, 51],
    'Bio' : [93, 41, 23,90, 45, 51],
    'Sanskrit' : [73, 81, 53, 98, 72, 21],
}
d1=pd.DataFrame(details)
d1
```

```python
d1['Total']=0
```

```python
d1
```

```python
d1['Total']=d1['Maths']+d1['Sci']+d1['Hindi']+d1['Chem']+d1['Bio']+d1['Sanskrit']
d1
```

```python
d1["Grade"]=0
```

```python
d1
```

```python
d1["%"]=(d1["Total"]/600)*100
```

```python
d1
```

less than 45-Fail

45-60 pass

60-70 first class

70-distinction

```python
d1.loc[d1["%"]<45]

d1.loc[d1["%"]<45,"Grade"]="FAIL"
```

```python
d1
```

```python
d1.loc[(d1["%"]>45) & (d1["%"]<60)]

d1.loc[(d1["%"]>45) & (d1["%"]<60),"Grade"]="PASS"
```

```python
(d1["%"]>45) & (d1["%"]<60)
```

```python
d1
```

```python
d1.loc[(d1["%"]>70)]
d1.loc[(d1["%"]>70),"Grade"]="DISTINCTION"

d1.loc[(d1["%"]>60) & (d1["%"]<70)]
d1.loc[(d1["%"]>60) & (d1["%"]<70),"Grade"]="First Class"
```
